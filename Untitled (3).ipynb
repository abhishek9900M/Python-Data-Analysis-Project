```python
df["State_order"] = np.random.choice(["Delhi", "Mumbai", "Chennai"], size=len(df))
```

```python
df.head()
```

```python
df.drop("New_Column", axis=1, inplace=True)
```

```python
df.head()
```

```python
def conv(value):
    if value =="Delhi":
        return "DEL"
    elif value =="Chennai":
        return "CHN"
    elif value =="Mumbai":
        return "Mum"
df["State_order"]= df["State_order"].apply(conv)

df.head(20)
```

```python
df.head()
```

```python
ax = sns.countplot(x="State_order",data = df )
ax.bar_label(ax.containers[0])
plt.show()
```

```python
gb = df.groupby("State_order").agg({"State_order": "count"})
plot=plt.pie(gb["State_order"],labels = gb.index, autopct= "%1.3f%%")
plt.title("State-Order-chart")
plt.show()
gb
```
